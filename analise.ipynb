```python
import yfinance as yf
```

```python
symbol = 'PETR4.SA'
```

```python
petra = yf.Ticker(symbol)
```

```python
type(petra)
```

```python
petra.info
```

```python
data = petra.history('1mo')
```

```python
data
```

```python
import streamlit as st
```

```python
atual = data.iloc[1,:]
atual
```

```python

close = atual.Close
delta = atual.Close - atual.Open 

st.title('Teste')
st.metric(
    label = 'Petr4.SA',
    value = close,
    delta = delta
)
```
